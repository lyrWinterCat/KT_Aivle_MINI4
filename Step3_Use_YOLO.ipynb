```python
import os

# label 폴더 경로 설정
labels_folder = 'class_face/label'

# label 폴더 내의 모든 텍스트 파일 처리
for filename in os.listdir(labels_folder):
    if filename.lower().endswith('.txt'):  # 텍스트 파일만 처리
        file_path = os.path.join(labels_folder, filename)

        # 파일 읽기
        with open(file_path, 'r') as file:
            lines = file.readlines()

        # 첫 번째 값을 1로 변경
        modified_lines = []
        for line in lines:
            parts = line.split()  # 공백을 기준으로 분리
            if parts:  # 비어있지 않은 경우
                parts[0] = '1'  # 0번째 값을 1로 변경
            modified_lines.append(' '.join(parts))  # 다시 문자열로 결합

        # 변경된 내용을 파일에 쓰기
        with open(file_path, 'w') as file:
            file.write('\n'.join(modified_lines))

print("파일 수정 완료!")
```

### (2) 데이터셋 전처리

* **세부 요구사항**
    - Label 파일, 즉 txt 파일의 클래스를 수정.
        - 여러 클래스로 구분되어 있는 것을 하나로 통일.

    - [UltraLytics YOLO 모델의 데이터셋 구조 링크](https://docs.ultralytics.com/datasets/detect/)
    - 예시 코드에서 사용한 라이브러리
        - os, glob, random, shutil, numpy

#### 1) Label 파일의 클래스 값 변경

```python
import os

# 경로 설정
base_path = '/content/other_face'  # your base path
subfolders = ['test', 'train', 'valid']

# 각 서브폴더의 labels 폴더 탐색
for subfolder in subfolders:
    labels_folder = os.path.join(base_path, subfolder, 'labels')

    # labels 폴더 내의 모든 .txt 파일 탐색
    for txt_file in os.listdir(labels_folder):
        if txt_file.endswith('.txt'):
            file_path = os.path.join(labels_folder, txt_file)

            # 파일 내용 읽기
            with open(file_path, 'r') as f:
                lines = f.readlines()

            # 첫 번째 값 변경
            new_lines = []
            for line in lines:
                parts = line.split()
                if parts:  # 공백이 아닌 경우
                    parts[0] = '1'  # 첫 번째 값을 1로 변경
                new_lines.append(' '.join(parts))

            # 변경된 내용으로 파일 덮어쓰기
            with open(file_path, 'w') as f:
                f.write('\n'.join(new_lines))

print("모든 .txt 파일의 첫 번째 값이 1로 변경되었습니다.")
```

#### 3) 각 폴더에 데이터 옮기기

```python
import os
import shutil
import random

# 경로 설정
my_face_path = '/content/my_face'  # my_face 폴더 경로
other_face_path = '/content/other_face'  # other_face 폴더 경로
output_path = '/content/dataset'  # 최종 YOLO 학습 구조 폴더 경로

# YOLO 학습 구조 생성
os.makedirs(os.path.join(output_path, 'images', 'train'), exist_ok=True)
os.makedirs(os.path.join(output_path, 'images', 'val'), exist_ok=True)
os.makedirs(os.path.join(output_path, 'images', 'test'), exist_ok=True)
os.makedirs(os.path.join(output_path, 'labels', 'train'), exist_ok=True)
os.makedirs(os.path.join(output_path, 'labels', 'val'), exist_ok=True)
os.makedirs(os.path.join(output_path, 'labels', 'test'), exist_ok=True)

# my_face에서 이미지와 라벨 복사
my_face_images = os.path.join(my_face_path, 'images')
my_face_labels = os.path.join(my_face_path, 'labels')

my_face_image_files = os.listdir(my_face_images)
my_face_label_files = os.listdir(my_face_labels)

# my_face 이미지 및 라벨을 랜덤하게 나누기
random.shuffle(my_face_image_files)
num_my_face = len(my_face_image_files)

train_my_face = my_face_image_files[:int(num_my_face * 0.7)]
val_my_face = my_face_image_files[int(num_my_face * 0.7):int(num_my_face * 0.85)]
test_my_face = my_face_image_files[int(num_my_face * 0.85):]

# my_face 파일 복사
for img_file in train_my_face:
    shutil.copy(os.path.join(my_face_images, img_file), os.path.join(output_path, 'images', 'train', img_file))
    shutil.copy(os.path.join(my_face_labels, img_file.replace('.jpg', '.txt')), os.path.join(output_path, 'labels', 'train', img_file.replace('.jpg', '.txt')))

for img_file in val_my_face:
    shutil.copy(os.path.join(my_face_images, img_file), os.path.join(output_path, 'images', 'val', img_file))
    shutil.copy(os.path.join(my_face_labels, img_file.replace('.jpg', '.txt')), os.path.join(output_path, 'labels', 'val', img_file.replace('.jpg', '.txt')))

for img_file in test_my_face:
    shutil.copy(os.path.join(my_face_images, img_file), os.path.join(output_path, 'images', 'test', img_file))
    shutil.copy(os.path.join(my_face_labels, img_file.replace('.jpg', '.txt')), os.path.join(output_path, 'labels', 'test', img_file.replace('.jpg', '.txt')))

# other_face에서 이미지 및 라벨 복사
other_face_subfolders = ['train', 'valid', 'test']

for subfolder in other_face_subfolders:
    other_face_images = os.path.join(other_face_path, subfolder, 'images')
    other_face_labels = os.path.join(other_face_path, subfolder, 'labels')

    # 이미지 및 라벨 파일 목록 가져오기
    other_face_image_files = os.listdir(other_face_images)
    other_face_label_files = os.listdir(other_face_labels)

    # 랜덤하게 나누기
    random.shuffle(other_face_image_files)
    num_other_face = len(other_face_image_files)

    train_other_face = other_face_image_files[:int(num_other_face * 0.7)]
    val_other_face = other_face_image_files[int(num_other_face * 0.7):int(num_other_face * 0.85)]
    test_other_face = other_face_image_files[int(num_other_face * 0.85):]

    # other_face 파일 복사
    for img_file in train_other_face:
        shutil.copy(os.path.join(other_face_images, img_file), os.path.join(output_path, 'images', 'train', img_file))
        shutil.copy(os.path.join(other_face_labels, img_file.replace('.jpg', '.txt')), os.path.join(output_path, 'labels', 'train', img_file.replace('.jpg', '.txt')))

    for img_file in val_other_face:
        shutil.copy(os.path.join(other_face_images, img_file), os.path.join(output_path, 'images', 'val', img_file))
        shutil.copy(os.path.join(other_face_labels, img_file.replace('.jpg', '.txt')), os.path.join(output_path, 'labels', 'val', img_file.replace('.jpg', '.txt')))

    for img_file in test_other_face:
        shutil.copy(os.path.join(other_face_images, img_file), os.path.join(output_path, 'images', 'test', img_file))
        shutil.copy(os.path.join(other_face_labels, img_file.replace('.jpg', '.txt')), os.path.join(output_path, 'labels', 'test', img_file.replace('.jpg', '.txt')))

print("YOLO 학습 구조가 생성되고 이미지가 복사되었습니다.")
```

    - [UltraLytics YOLO 추론 명령어 링크](https://docs.ultralytics.com/modes/predict/)

```python
import os
import random
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# YOLO 모델 로드 (훈련 후 저장된 모델 사용)
model = YOLO("runs/detect/train/weights/best.pt")

# 이미지가 있는 폴더 경로 설정
my_face_dir = "/content/my_face/images"  # my_face 폴더 경로
other_face_dir = "/content/class_face/img"  # other_face 폴더 경로

# 두 개의 폴더에서 이미지 파일 리스트 가져오기
my_face_images = [os.path.join(my_face_dir, f) for f in os.listdir(my_face_dir) if f.endswith(('.png', '.jpg', '.jpeg'))]
other_face_images = [os.path.join(other_face_dir, f) for f in os.listdir(other_face_dir) if f.endswith(('.png', '.jpg', '.jpeg'))]

# 모든 이미지 리스트 결합
all_images = my_face_images + other_face_images

# 랜덤으로 이미지 선택
random_image_path = random.choice(all_images)

# 이미지 읽기
image = cv2.imread(random_image_path)

# 이미지가 제대로 읽혔는지 확인
if image is None:
    print(f"Error: Unable to read image at {random_image_path}. Please check the file path.")
else:
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # OpenCV는 BGR 포맷으로 이미지를 읽기 때문에 RGB로 변환

    # 이미지 추론
    results_inference = model.predict(image)

    # 결과 출력
    for result in results_inference:
        print("Original Shape:", result.orig_shape)  # 원본 이미지 크기
        print("Image Path:", result.path)  # 이미지 경로
        print("Processing Speed:", result.speed)  # 처리 속도

        # 확률 확인
        if hasattr(result, 'probs') and result.probs is not None:
            # result.probs가 리스트인지 확인
            if isinstance(result.probs, list):
                for class_id, prob in enumerate(result.probs):
                    print(f"Class ID: {class_id}, Probability: {prob:.4f}")
            else:
                print("Probability data is not in expected format.")
        else:
            print("No probability data available.")

        # 결과 시각화
        plt.figure(figsize=(10, 10))
        plt.imshow(result.plot())  # 예측 결과 시각화
        plt.axis('off')
        plt.title(f'Predicted: {result.path}')
        plt.show()
```

```python
import glob
import numpy as np
import torch
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# YOLO 모델 로드
model = YOLO("yolo11n.pt")

# 이미지 파일 경로 리스트
img_fp_list = glob.glob('/content/dataset/images/*/*.jpg')

# 예측 결과를 저장할 리스트
predictions = []

# 모델 추론
for img_fp in img_fp_list:
    # 이미지 읽기
    img = cv2.imread(img_fp)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV는 BGR 포맷이므로 RGB로 변환

    # 예측 수행
    results = model(img)

    # 결과 저장
    predictions.append(results)

    # 결과 시각화
    # 바운딩 박스를 그리기 위해 결과를 시각화합니다.
    img_with_detections = img.copy()
    results[0].plot()  # 첫 번째 이미지에 대한 결과를 플롯합니다.

    # plt.imshow(img_with_detections)
    plt.axis('off')
    plt.show()

# 예측 결과 확인
for i, result in enumerate(predictions):
    # print(f"Image: {img_fp_list[i]}")
    print("Predicted boxes:", result[0].boxes.xyxy)  # 예측된 바운딩 박스 정보
    print("Predicted classes:", result[0].boxes.cls)  # 예측된 클래스 정보
    print("Predicted confidences:", result[0].boxes.conf)  # 신뢰도 정보
```
